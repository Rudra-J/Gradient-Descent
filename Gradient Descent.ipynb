```python
#gedient descent my linal
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(5)
x=np.array([[5,3],[6,4],[7,6],[9,8]])
y=np.array([2,4,7,8]).reshape(-1,1)
```

```python
def grad_desc_linal(x,y,lr,itter):
    """x-> {independet var}array type input, y-> {dependetn var}array type input, itter->{number of itterations}int input"""
    m,n,lr=x.shape[0],x.shape[1],0.01
    w,b=np.zeros(shape=[n,1]),0  # w=np.random.rand(n,1)
    cost=[]
    for i in range(itter):    
        h=np.dot(x,w)+b
        j=(1/(2*m))*np.sum((h-y)**2)
        dw=np.dot(x.T,h-y)/m
        db=np.sum(h-y)/m
        w-=lr*dw
        b-=lr*db
        cost.append(j)
    plt.plot(cost)
    return w,b
```

```python
grad_desc_linal(x,y,0.01,100000)
```

```python
np.random.seed(123)
x=np.random.rand(100,8)
y=np.random.rand(100,1).reshape(-1,1)
```

```python
grad_desc_linal(x,y,0.01,100000)
```
